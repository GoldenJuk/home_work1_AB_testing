## Задание 1

***ozon .ru***

*Какие бы ключевые метрики вы выделили? Опишите почему вы выбрали именно их и можно ли их измерить*

1. <u>*Конверсия.*<u>

Этот показатель измеряет процент посетителей, которые перешли на сайт и совершили покупку. Эта метрика нужна для оценки эффективности маркетинга.
    
2. <u>*Средняя стоимость заказа*.<u> 
    
Эта мерика показывает среднюю сумму, потраченную на заказ. Увеличение этого показателя показывает, что клиенты довольны продуктами.
    
3. <u>*Количество пользователей*.<u>
    
Эта ключевая метрика, т.к. она напрямую влияет на выручку.
    
4. <u>*Стоимость привлечения клиента.*<u> 
    
Эта метрика показывает стоимость привлечения нового клиента. Необходимо стремиться к минимизации этого показателя.
    
5. <u>*Коэффициент удержания клиентов.*<u>
    
Этот показатель измеряет процент клиентов, которые возвращаются, чтобы совершить покупку снова. Высокий показатель удержания клиентов указывает на лояльность клиентов.

## Задание 2

*Сформулируйте 5-10 гипотез для выбранного продукта.*

1. Улучшив пользовательский интерфейс продукта, мы можем увеличить количество пользователей, прошедших процесс регистрации.
2. Увеличение количества товаров в каталоге на 50% увеличит количество конверсий на 10%.
3. Уменьшение количества дней доставки среднестатистического заказа до 2 дней повышает лояльность покупателей и увеличивает повторные покупки на 10%.
4. Увеличение инвестиций в рекламу в социальных сетях на 30% приведет к увеличению конверсии на 5%.
5. Добавляя такие элементы, как отзывы и рейтинги клиентов, мы можем повысить доверие потенциальных клиентов и побудить их совершить покупку.
6. Предлагая программу лояльности или реферальную программу, мы можем увеличить удержание существующих клиентов.
7. Создание программы персонализации на основе истории покупок приведет к увеличению средней выручки на 10%.
8. Предлагая различные способы оплаты, мы можем снизить количество незавершенных заказов и увеличить конверсию.
